# Take 2 images and combine it to form a single image..

```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
pht1=cv2.imread('scenery1.jpg')
```

```python
pht2=cv2.imread('scenery2.jpg')
```

```python
cv2.imshow("pht1",pht1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow("pht2",pht2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
b=pht2
```

```python
b.shape
```

```python
a.shape
```

```python
# now both photos have same rows and columns
# vertical stack

v_stack  = numpy.vstack((a,b))

plt.imshow(v_stack)

cv2.imshow("Vertical stack",v_stack)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# horizontal stack

h_stack  = numpy.hstack((a,b))

plt.imshow(h_stack)

cv2.imshow("Horizontal stack",h_stack)
cv2.waitKey()
cv2.destroyAllWindows()
```
